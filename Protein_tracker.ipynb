Daily Protein & Calories Tracker

```python
def add_food():
    food = input("Enter food name: ")
    protein = float(input("Protein (in grams): "))
    calories = float(input("Calories (kcal): "))

    with open("nutrition_log.csv", "a") as file:
        file.write(f"{food},{protein},{calories}\n")

    print("Saved ✅")


def show_day_summary():
    total_protein = 0
    total_calories = 0

    print("\nToday's Summary:\n")

    with open("nutrition_log.csv", "r") as file:
        for line in file:
            food, protein, calories = line.strip().split(",")
            total_protein += float(protein)
            total_calories += float(calories)
            print(f"{food} → Protein: {protein}g, Calories: {calories} kcal")

    print("\n----------------------")
    print(f"Total protein: {total_protein} g")
    print(f"Total calories: {total_calories} kcal")
    print("----------------------\n")


while True:
    print("\n1) Add food")
    print("2) Show today's summary")
    print("3) Exit")

    choice = input("Choose: ")

    if choice == "1":
        add_food()
    elif choice == "2":
        show_day_summary()
    else:
        break
```
